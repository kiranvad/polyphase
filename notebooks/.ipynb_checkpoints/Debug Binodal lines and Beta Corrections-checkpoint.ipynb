```python
%config InlineBackend.figure_format = 'svg'
import mpltern
from matplotlib import rc
rc('text', usetex=True)
import seaborn as sns
import matplotlib.pyplot as plt

import pdb
import numpy as np

import sys
if '../' not in sys.path:
    sys.path.append('../')

from solvers.utils import makegrid3d, scatter3d, contours,plotlocus

%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

from plotly.subplots import make_subplots
from solvers.utils import scatter3d
from solvers.helpers import get_ternary_coords
import plotly.graph_objects as go
```

```python
grid = np.asarray(makegrid3d(num=40))
```

```python
chi = [0.341, 0.885, 0.941]
CHI = np.array([[0,chi[0],chi[1]],[chi[0],0,chi[2]],[chi[1],chi[2],0]])
M = np.array([  1, 66,   5])
M,CHI
```

```python
from solvers import cem,spinodal, helpers
gmix_beta = lambda x: helpers.flory_huggins(x, M, CHI,beta=1e-4)

energy_beta = []
for point in grid:
    energy_beta.append(gmix_beta(point))
    
gmix_nobeta = lambda x: helpers.flory_huggins(x, M, CHI,beta=0.0)

energy_nobeta = []
for point in grid:
    energy_nobeta.append(gmix_nobeta(point)) 
    
from solvers.helpers import log, FHTaylor
assert np.isclose(np.log(0.2),log(0.2,n=100))

gmix_FH_beta = lambda x: helpers.FHTaylor(x, M, CHI,beta=1e-4)

energy_FHbeta = []
for point in grid:
    energy_FHbeta.append(gmix_FH_beta(point)) 
    
energy_FH_nobeta = []
for point in grid:
    energy_FH_nobeta.append(helpers.FHTaylor(point, M, CHI,beta=0.0))  
```

```python
fig = plt.figure(figsize=(10.8, 4.8))
fig.subplots_adjust(left=0.075, right=0.85, wspace=0.65)
plt.set_cmap('viridis')

ax = fig.add_subplot(1,1,1,projection='ternary',rotation=-120)
diff_energy = (np.asarray(energy_beta)-np.asarray(energy_nobeta))/np.asarray(energy_nobeta)
ax,cs, cax = contours(grid,diff_energy,mode='contour',ax=ax)
pad_title = 36
ax.set_title("Energy Difference", pad=pad_title)
colorbar = fig.colorbar(cs, cax=cax)

plt.show()
```

```python
specs = [[{'type': 'scene'},{'type': 'scene'},{'type': 'ternary'}]]
fig = make_subplots(rows=1,cols=3,specs=specs,subplot_titles=("Beta corrected Energy",\
                                                              "W/O beta correction","Energy Difference"))
coords = []
for point in grid:
    coords.append(get_ternary_coords(point))
    
trace = scatter3d(coords,energy_beta)
fig.add_trace(trace, row=1, col=1)
fig.update_layout(scene = dict(zaxis_title='Energy'))
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))

trace = scatter3d(coords,energy_nobeta)
fig.add_trace(trace, row=1, col=2)
fig.update_layout(scene = dict(zaxis_title='Energy'))
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))

def makeAxis(title, tickangle):
    return {
      'title': title,
      'titlefont': { 'size': 10 },
      'tickangle': tickangle,
      'tickfont': { 'size': 15 },
      'tickcolor': 'rgba(0,0,0,0)',
      'ticklen': 5,
      'showline': True,
      'showgrid': True
    }

trace = go.Scatterternary({
    'mode': 'markers',
    'a': np.asarray(grid)[:,0],
    'b': np.asarray(grid)[:,1],
    'c': np.asarray(grid)[:,2],
    'hovertemplate' : '<b>Difference: %{text:.2f}</b>',
    'text':diff_energy
})

fig.add_trace(trace, row=1, col=3)
fig.update_layout({
    'ternary': {
        'sum': 100,
        'aaxis': makeAxis('Phi_s', 0),
        'baxis': makeAxis('Phi_p1', 45),
        'caxis': makeAxis('Phi_p2', -45)
    }
})
fig.write_html("../expts/data/betacorrectioncompare.html")
#fig.show()
```

```python
specs = [[{'type': 'scene'},{'type': 'scene'}],[{'type': 'scene'},{'type': 'scene'}]]
fig = make_subplots(rows=2,cols=2,specs=specs,subplot_titles=("Beta corrected Energy",\
                                                              "W/O beta correction",\
                                                              "FH Taylor Series beta","FH Taylor Series w/o beta"))
coords = []
for point in grid:
    coords.append(get_ternary_coords(point))
    
trace = scatter3d(coords,energy_beta)
fig.add_trace(trace, row=1, col=1)
fig.update_layout(scene = dict(zaxis_title='Energy'),scene_aspectmode='cube')

trace = scatter3d(coords,energy_nobeta)
fig.add_trace(trace, row=1, col=2)
fig.update_layout(scene = dict(zaxis_title='Energy'),scene2_aspectmode='cube')

trace = scatter3d(coords,energy_FHbeta)
fig.add_trace(trace, row=2, col=1)
fig.update_layout(scene = dict(zaxis_title='Energy'),scene3_aspectmode='cube')

trace = scatter3d(coords,energy_FH_nobeta)
fig.add_trace(trace, row=2, col=2)
fig.update_layout(scene = dict(zaxis_title='Energy'),scene4_aspectmode='cube')

fig.update_layout(
    title_text='Comparing tayor series and beta correction',
    height=800,
    width=800, margin=dict(l=80, r=80, t=100, b=80)
)


fig.write_html("../expts/data/FHComparision.html")

#fig.show()
```

```python
# Plot a convex triangulation in the ternary space
from scipy.spatial import ConvexHull
coords = []
for point in grid:
    coords.append(get_ternary_coords(point))
points = np.concatenate((coords,np.asarray(energy_beta).reshape(-1,1)),axis=1)
hull = ConvexHull(points)
coords = np.asarray(coords)
fig1, ax1 = plt.subplots()
ax1.set_aspect('equal')
ax1.triplot(coords[:,0], coords[:,1], triangles= hull.simplices, lw=1)
ax1.set_title('Triangulation with convex hull')
plt.axis('off')
plt.show()
```
