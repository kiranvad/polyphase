```python
import mpltern
from matplotlib import rc
rc('text', usetex=True)
import seaborn as sns
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap
plt.rcParams.update({'font.size': 20})

import pdb
import numpy as np

import sys
if './' not in sys.path:
    sys.path.append('./')

from phasemodelling import twod_precomputed, phase
from utils import makegrid2d
```

```python
grid = np.asarray(makegrid2d(num=100))
```

```python
M = np.array([1,1])
chisp = 3
CHI = np.array([[0,chisp],[chisp,0]])
M,CHI
```

```python
curvature,energy,pcspinodal, pcenergy = [],[],[],[]

temp = []
twodpc = twod_precomputed(CHI,M)
p = phase(CHI,M)

for point in grid:
    energy.append(p.free_energy(point))
    pcenergy.append(twodpc.freeenergy(point))
    curv, Hag, eigs = p.spinodal(point)
    curvature.append(curv)
    Hpc = twodpc.hessian(point)
    label, value = twodpc.spinodal(point)
    pcspinodal.append(label)
    temp.append(eigs)
    
    assert np.allclose(Hag, Hpc)
```

```python
fig = plt.figure(figsize=(10.8, 4.8))
fig.subplots_adjust(left=0.075, right=0.85, wspace=0.65)

ax = fig.add_subplot(1,2,1)
p = ax.scatter(grid[:,0],energy, c = curvature, cmap='bwr')
ax.set_ylabel('Free Energy Matrix')
fig.colorbar(p,ax=ax)
ax.set_title('Spinodal using eigen values',fontsize=15)
ax = fig.add_subplot(1,2,2)
ax.scatter(grid[:,0],pcenergy,c = pcspinodal, cmap='bwr')
#ax.set_ylabel('Free Energy Pre-computed')
ax.set_title('Spinodal using curvature = 1',fontsize=15)
sns.despine()
fig.suptitle('Solution of a 2D System', y=1.05)
fig.text(0.5, -0.01, '$\\varphi_1$', va='center', fontsize=20)

plt.tight_layout()
#plt.savefig('twod_spinodal_example.pdf',dpi=500,bbox_inches='tight')
plt.show()
```

```python
# Convex Hull based method to find binodal curve
from scipy.spatial import ConvexHull
points = np.concatenate((grid[:,1].reshape(-1,1),np.asarray(energy).reshape(-1,1)),axis=1)
hull = ConvexHull(points)
```

```python
plt.plot(points[:,0], points[:,1], 'o')
for simplex in hull.simplices:
    plt.plot(points[simplex, 0], points[simplex, 1], 'k-')
    
plt.show()
```
