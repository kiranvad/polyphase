```python
import ipywidgets as widgets
from math import pi
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt

def on_value_change(change):
    execute()

period = widgets.IntSlider(min=-5,max=5,value=1,description='Period:',continuous_update=False)
period.observe(on_value_change, names='value')

scale = widgets.FloatSlider(min=0.1,max=5,step=0.1,value=1,description='Scale:' ,continuous_update=False)
scale.observe(on_value_change, names='value')

display(widgets.HBox(( period,scale)))

plt.figure()
def execute():
    plt.clf()
    p = period.value
    s = scale.value
    
    x = np.linspace(-p*pi,p*pi)
    y = np.sin(x)
    ax = plt.subplot()
    ax.plot(x,y)
    plt.tight_layout()
    plt.show()
    
execute()
```
