## Checking if something from chemical potentials makes sense
This is just way too diffucult to get any intuition

```python
def compute_chemical_potential(phi,m,chi):
    mu1 = (phi[1]**2)*chi[0] + chi[1]*(phi[2]**2) + \
    phi[2]*(1-(1/m[2]) + phi[1]*(chi[0]+chi[1]-chi[2])) + np.log(phi[0])
    
    mu2 = chi[0]*(phi[1]-1)**2 + chi[1]*phi[2]**2 - phi[2]/m[2] + \
    phi[2]*((1 + (phi[1]-1))*(chi[0]+chi[1])+chi[2] - phi[1]*chi[2]) + np.log(phi[1])
    
    mu3 = 1 - phi[2] + m[2]*(-1 + chi[1] + chi[1]*phi[2]**2) + \
    m[2]*(phi[2]*(1-2*chi[1]+phi[1]*(chi[0] + chi[1]-chi[2])) + phi[1]*(chi[0]*(phi[1]-1)-chi[1] + chi[2])) + np.log(phi[2])
    
    return np.array([mu1,mu2,mu3])
```

```python
criteria = (np.asarray(num_comps)==2).reshape(-1)
simplex_vertices = np.unique(simplices[criteria,:])
chempots = [compute_chemical_potential(grid[:,x],configuration['M'],configuration['chi']) for x in simplex_vertices]
chempots = np.asarray(chempots)
simplex_coords = grid[:,simplex_vertices]
fig = plt.figure(figsize=(3*4*1.6,4))
fig.subplots_adjust(wspace=0.2)
for i in range(chempots.shape[1]):
    ax = fig.add_subplot(1,3,i+1, projection='ternary')
    pc = ax.scatter(simplex_coords[2,:], simplex_coords[0,:],\
                    simplex_coords[1,:], c=chempots[:,i])
    ax.set_tlabel('solvent')
    ax.set_llabel('polymer')
    ax.set_rlabel('small molecule')
    ax.taxis.set_label_position('tick1')
    ax.laxis.set_label_position('tick1')
    ax.raxis.set_label_position('tick1')
    cax = ax.inset_axes([1.05, 0.1, 0.05, 0.9], transform=ax.transAxes)
    colorbar = fig.colorbar(pc, cax=cax)
    
colorbar.set_label('Chemical potenital', rotation=270, va='baseline')
```

# Plotting simplices and points inside it

Tells us how the labels are assigned for each point in a uniform grid from simplex. This method was used to identify the problem with spurious 3-phase regions and traced it back to coplanar simplices in the phi space

```python
from solvers import helpers
from solvers.visuals import _set_axislabels_mpltern
from parallel.parphase import makegridnd

def is_collinear(coords):
    """ returns True if three points are collinear """
    M= np.vstack((coords.T,np.array([1,1,1])))
    area = np.linalg.det(M)    
    flag = area<1e-15
    
    return flag

def show_chempots(simplex_tri_coords,simplex_chempots):
    fig = plt.figure(figsize=(3*4*1.6,4))
    fig.subplots_adjust(wspace=0.2)
    for i in range(3):
        ax = fig.add_subplot(1,3,i+1)
        ax.add_patch(plt.Polygon(np.array([[0,0],[1,0],[0.5,0.866]]), facecolor='none', edgecolor='k', lw=1.0))
        ax.scatter(simplex_tri_coords[:,0],simplex_tri_coords[:,1],c='k')
        s = np.array(['{:.2f}'.format(x[i]) for x in simplex_chempots])
        for j, txt in enumerate(s):
            ax.annotate(txt, (simplex_tri_coords[j,0], simplex_tri_coords[j,1]))
        ax.add_patch(plt.Polygon(simplex_tri_coords))
        ax.set_title(r'$\mu_{}$'.format(i+1))
        ax.axis('off')
 
    return ax, fig

def show_simplex_mpltern(euclidean_coords,ax):
    fig = plt.gcf()
    ax.fill(euclidean_coords[:,2], euclidean_coords[:,0], euclidean_coords[:,1], facecolor='b', alpha=0.75)
    ax.scatter(euclidean_coords[:,2], euclidean_coords[:,0], euclidean_coords[:,1],c='k', s = 10, alpha=0.5)
    _set_axislabels_mpltern(ax)


def show_collinear_simplex(euclidean_coords,ax):
    fig = plt.gcf()
    ax.scatter(euclidean_coords[:,2], euclidean_coords[:,0], euclidean_coords[:,1],c='k', s =10, alpha=0.5)
    _set_axislabels_mpltern(ax)
    
    return ax, fig

from solvers.helpers import inpolyhedron
def plot_points_inside(grid, lift_grid, simplex, ax):
    fig = plt.gcf()
    v = np.asarray([grid[:-1,x] for x in simplex])
    inside = inpolyhedron(v, lift_grid[:-1,:].T)
    euclidean_coords = lift_grid.T
    ax.scatter(euclidean_coords[inside,2], euclidean_coords[inside,0], euclidean_coords[inside,1],c='b',s = 2, alpha=0.5)
    _set_axislabels_mpltern(ax)
    
    return ax, fig    
```

```python
label = 3
inds = np.argwhere(np.asarray(num_comps)==label)
plt.close('all')
collinear = []
lift_grid = makegridnd(200, 3)
tri_coords = np.array([helpers.get_ternary_coords(pt) for pt in grid.T])
plt.close('all')
for ind in inds:
    ind = np.array([0])
    triangle = simplices[ind[0]]
    simplex_chempots = np.array([compute_chemical_potential(grid[:,x],configuration['M'],configuration['chi']) for x in triangle])
    simplex_tri_coords = np.array([tri_coords[x,:] for x in triangle])
    simplex_euclidean_coords = np.array([grid[:,x] for x in triangle])
    flag = is_collinear(simplex_tri_coords)
    fig, ax = plt.subplots(subplot_kw={'projection':'ternary'})
    
    if flag:
        ax, fig = show_collinear_simplex(simplex_euclidean_coords, ax)
    else:
        ax, fig = show_simplex_mpltern(simplex_euclidean_coords, ax)
        #ax, fig = plot_points_inside(grid,lift_grid, triangle, ax)  
        
    fig.suptitle('Simplex :{} is Collinear : {} and {}-Phase'.format(ind,flag, label))
#     fname = '../figures/debugliftlabel/{}phase_iscollinear_{}_simplexid_{}'.format(label,flag, ind[0])
#     plt.savefig(fname, dpi=300, bbox_inches='tight')
#     plt.close()
    break
plt.show()
```

## Understand the simplicial topology 

Some of the simplices near the boundaries can be sometimes mis-classified. One way to check is if they've same same topology as their ideal counterparts.

```python
import cechmate as cm
from persim import plot_diagrams
rips = cm.Rips(maxdim=1, verbose=False) #Go up to 1D homology
```

```python

def plot_simplicial_topology(euclidean_coords,ax):
    rips.build(euclidean_coords)
    dgmsrips = rips.diagrams()
    plot_diagrams(dgmsrips, ax=ax)
    ax.set_title("Rips Persistence Diagrams")
    
    return dgmsrips[0]

def get_H0curve(euclidean_coords):
    rips.build(euclidean_coords)
    dgmsrips = rips.diagrams()
    H0 = dgmsrips[0]
    cps = np.hstack((0.0,np.sort(H0[:,1]), 1.5))
    eps = 1e-5
    y = [np.sum(cps>x-eps) for x in cps]
    y[0] = 3
    y[-1] = 1
    
    x = cps.tolist()

    return x, y
```

```python
# Visualize simplex and its persistent homology
from time import sleep
from IPython.display import clear_output, display

dirname = '../figures/simplicial_topology/'
import os
if not os.path.exists(dirname):
    os.makedirs(dirname)

fullrun = False
label = 2
inds = np.argwhere(np.asarray(num_comps)==label)
print('Total of {} simplices with {} phase'.format(len(inds), label))
for i, ind in enumerate(inds):
    clear_output(wait=True)
    print("Processing {}/{}".format(i,len(inds)), end="", flush=True)
    fig = plt.figure(figsize=(3*3*1.6, 3))
    triangle = simplices[ind[0]]
    simplex_euclidean_coords = np.array([grid[:,x] for x in triangle])
    ax0 = fig.add_subplot(1,3,1, projection='ternary')
    show_simplex_mpltern(simplex_euclidean_coords, ax0)
    
    ax1 = fig.add_subplot(1,3,2)
    #plot_simplicial_topology(simplex_euclidean_coords,ax1)
    x, y = get_H0curve(simplex_euclidean_coords)

    ax1.step(x,y)
    ax1.scatter(x,y)
    ax1.set_yticks([1, 2, 3])
    ax1.set_ylabel('Connected Components')
    ax1.set_xlabel('Distance threshold')
    
    ax = fig.add_subplot(1,3,3)
    H0 = plot_simplicial_topology(simplex_euclidean_coords,ax)
    
    title = r'{} phase index {}'.format(num_comps[ind[0]], ind[0])
    fig.suptitle(title)
    fname = dirname + title + '.png'
    if fullrun:
        plt.savefig(fname, bbox_inches='tight')
        plt.close('all')
    else:
        plt.show()
        break
    sleep(0.01)
```

```python
# label simplex based on betticurve
from scipy.interpolate import interp1d, PchipInterpolator
x, y = get_H0curve(simplex_euclidean_coords)
f_interp = interp1d(x, y, kind='next')
x_new = np.linspace(0, np.max(x)-1e-5, num=11, endpoint=True)
y_new = f_interp(x_new)

fig, ax = plt.subplots()
ax.scatter(x,y, label='0-Betti curve')
ax.plot(x_new, y_new, label = 'interpolation')
ax.set_yticks([1, 2, 3])
ax.set_ylabel('Connected Components')
ax.set_xlabel('Distance threshold')
ax.legend()
plt.show()
```
